```python
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(s: str) -> str:
    # a) to string, lowercase and strip
    s = str(s).lower().strip()

    # b) expand common contractions
    s = contractions.fix(s)

    # c) remove URLs
    s = re.sub(r'https?://\S+|www\.\S+', '', s)

    # d) remove @mentions and #hashtags
    s = re.sub(r'@\w+|#\w+', '', s)

    # e) remove punctuation (keep whitespace and word‐chars)
    s = re.sub(r'[^\w\s]', ' ', s)

    # f) remove emojis and other symbols
    s = emoji.replace_emoji(s, replace='')  # requires `emoji` package

    # g) tokenize on whitespace
    tokens = s.split()

    # h) lemmatize each token
    tokens = [lemmatizer.lemmatize(tok) for tok in tokens]

    # i) re‐join into a cleaned string (or return tokens list if you prefer)
    return ' '.join(tokens)
```

```python
mydata['cleaned text'] = mydata['text'].apply(clean_text)
```

```python
def avg_word2vec1(sentence, model):
    words = sentence.lower().split()
    valid_vectors = [model[word] for word in words if word in model]
    
    if not valid_vectors:
        return np.zeros(model.vector_size)
    
    return np.mean(valid_vectors, axis=0)
```
